## Sales Peak During the Year

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

monthly_sales = df.groupby([pd.Grouper(key='Order Date', freq='M'), 'Product Name'])['Sales'].sum()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

monthly_sales = df.groupby([
    pd.Grouper(key='Order Date', freq='M'),
    'Product Name'
])['Sales'].sum().reset_index()
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().nlargest(20).index
monthly_sales_top = monthly_sales[monthly_sales['Product Name'].isin(top_products)]
```

```python
pivot = monthly_sales_top.pivot(index='Order Date', columns='Product Name', values='Sales')
```

```python
pivot.plot(figsize=(12,6))
plt.title('Monthly Sales Trend of Top 5 Products')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend(title='Product Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
monthly_sales = df.groupby(pd.Grouper(key='Order Date', freq='M'))['Sales'].sum()

import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
monthly_sales.plot(marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from scipy.signal import find_peaks

sales_values = monthly_sales.values
peaks, _ = find_peaks(sales_values)

# Plot with peaks
plt.figure(figsize=(12,5))
plt.plot(monthly_sales.index, sales_values, marker='o', label='Monthly Sales')
plt.plot(monthly_sales.index[peaks], sales_values[peaks], "ro", label='Peaks')
plt.title("Sales Peaks During the Year")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Which categories or regions bring the most revenue?

```python
#Sales by Region per Month
monthly_region_sales = df.groupby([
    pd.Grouper(key='Order Date', freq='M'),
    'Region'
])['Sales'].sum().reset_index()

#Sales by Category per Month
monthly_category_sales = df.groupby([
    pd.Grouper(key='Order Date', freq='M'),
    'Category'
])['Sales'].sum().reset_index()
```

```python
# For Region
pivot_region = monthly_region_sales.pivot(index='Order Date', columns='Region', values='Sales')

# For Category
pivot_category = monthly_category_sales.pivot(index='Order Date', columns='Category', values='Sales')
```

```python
#Plot Trends and Spot Peaks for Regions
pivot_region.plot(figsize=(12,6), marker='o')
plt.title("Monthly Sales Trends by Region")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend(title="Region", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
#Plot Trends and Spot Peaks for Categories
pivot_category.plot(figsize=(12,6), marker='o')
plt.title("Monthly Sales Trends by Category")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend(title="Category", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
from scipy.signal import find_peaks

for col in pivot_region.columns:
    sales = pivot_region[col].dropna()
    peaks, _ = find_peaks(sales.values)
    
    plt.figure(figsize=(10,4))
    plt.plot(sales.index, sales.values, marker='o', label=col)
    plt.plot(sales.index[peaks], sales.values[peaks], "ro", label='Peaks')
    plt.title(f"Sales Peaks - {col} Region")
    plt.xlabel("Month")
    plt.ylabel("Sales")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
